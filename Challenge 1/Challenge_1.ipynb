```python
sales = {}
temp = {}
thi = {}
staff = {}
```

```python
sales['mean'] = data['sales'].mean()
sales['var'] = data['sales'].var()
sales['min'] = data['sales'].min()
sales['max'] = data['sales'].max()
```

```python
temp['mean'] = data['temperature'].mean()
temp['var'] = data['temperature'].var()
temp['min'] = data['temperature'].min()
temp['max'] = data['temperature'].max()
```

```python
thi['mean'] = data['THI'].mean()
thi['var'] = data['THI'].var()
thi['min'] = data['THI'].min()
thi['max'] = data['THI'].max()
```

```python
staff['mean'] = data['num_staff'].mean()
staff['var'] = data['num_staff'].var()
staff['min'] = data['num_staff'].min()
staff['max'] = data['num_staff'].max()
```

```python
data['sales'].fillna(sales['mean'], inplace = True)
data['temperature'].fillna(temp['mean'], inplace = True)
data['THI'].fillna(thi['mean'], inplace = True)
data['num_staff'].fillna(staff['mean'], inplace = True)
data['weather'].fillna(1, inplace = True)

data
```

```python
def norm(x, min, max):
    y = (x - min)/(max-min)
    return y


```

```python
print(norm(data['temperature'], temp['min'], temp['max']))
```

```python
data['temperature'] = norm(data['temperature'], temp['min'], temp['max'])
data['THI'] = norm(data['THI'], thi['min'], thi['max'])
data['num_staff'] = norm(data['num_staff'], staff['min'], staff['max'])
```
